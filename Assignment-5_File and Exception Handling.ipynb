# 2 Write a Python program to copy the contents of one file to another file

```python
# Ask the user for source and destination file names
source_file = input("Enter the name of the source file:")
destination_file = input("Enter the name of the destination file:")

try:
    # Open source file in read mode
    with open(source_file, 'r') as src:
        content = src.read()

    # Open destination file in write mode and write the content
    with open(destination_file, 'w') as dest:
        dest.write(content)

    print("File copied successfully!")

except FileNotFoundError:
    print("The source file does not exist. Please check the file name and try again.")

except Exception as e:
    print("An error occurred:", e)
    
```

# 3 Write a Python program to read the content of a file and count the total number of words in that file.


```python
# prompt to enter the file name
file_name = input("Enter the file name : ")

try:
    # Open the file in read mode
    with open(file_name, 'r') as file:
        content = file.read()  # it read the full content

        words = content.split()  # it split the content into words
        word_count = len(words)  # it count the number of words

    print("Total number of words in the file:", word_count)

except FileNotFoundError:
    print("The file does not exist. Please check the file name and try again.")

except Exception as e:
    print("An error occurred:", e)
```

# 4 Write a Python program that prompts the user to input a string and converts it to an integer. Use try-except blocks to handle any exceptions that might occur

```python
# Prompt for user input
user_input = input("Enter a number: ")

try:
    # to convert the input to an integer
    number = int(user_input)
    print("The integer value is:", number)

except ValueError:
    # if conversion fails
    print("Invalid input. Please enter a valid number.")
    
```

# 5 Write a Python program that prompts the user to input a list of integers and raises an exception if any of the integers in the list are negative.

```python
def check_for_negatives(numbers):
    for num in numbers:
        if num < 0:
            raise ValueError("Negative value found")
```

```python
# Prompt user for input
try:
    user_input = input("Enter a list of integers separated by spaces: ")
    
    # Converting the input string to a list of integers
    int_list = [int(num) for num in user_input.split()]
    
    # Check for negative numbers
    check_for_negatives(int_list)
    
    print("All numbers are non-negative:", int_list)

except ValueError as e:
    print("Error:", e)
```

# 6  Python Program to Compute Average with Exception Handling and Finally Block

```python
try:
    # Get input from the user
    user_input = input("Enter a list of integers (separated by spaces): ")
    
    # Convert the input into a list of integers
    int_list = [int(num) for num in user_input.split()]
    
    # Check if the list is empty
    if len(int_list) == 0:
        raise ValueError("The list is empty. Cannot compute average.")
        
         # Compute average
    average = sum(int_list) / len(int_list)
    print("The average is:", average)

except ValueError as ve:
    print("Error:", ve)

except Exception as e:
    print("An unexpected error occurred:", e)

finally:
    print("Program has finished running.")
```

# 7 Python program that prompts the user to input a filename and writes a string to that file. Use try-except blocks to handle any exceptions that might occur and print a welcome message if there is no exception occurred.

```python

try:
    # Prompt the user for a filename
    filename = input("Enter the filename to write to: ")
    
    # Prompt the user for the content to write
    content = input("Enter the text you want to write to the file: ")
    
    # Open the file in write mode to write the content
    with open(filename, 'w') as file:
        file.write(content)
    
    # If no exception
    print("Congratulations! Your content has been successfully written to the file.")

except Exception as e:
    # Print any error that occurs
    print("An error occurred:", e)
```
